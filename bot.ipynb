**Librairies et fonctions**

```python
import faker
import random
import unidecode
from datetime import datetime
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
import time
import csv
import datetime

#-----Fonctions personnalisées

#Créé un numéro français aléatoire
def fakephone():
    phone_number = f"06{str(random.randint(10**9, (10**10)-1))}"
    return phone_number

#Associe une civilité à un prénom donné
#Civilité 1 si prénom non trouvé
def get_civilite(liste_prenoms_homme, prenom):
    prenom = unidecode.unidecode(prenom.lower(),"utf-8")
    civilite = 1
    if prenom not in liste_prenoms_homme:
        civilite = 2  
    return civilite
```

**Vote pour Marie**

```python
# Variables globales
#------------------------------------------------------------------------------------------------------
fake = Faker(locale="fr_FR") #Restreint aux consonnances françaises
domain_list = "@orange.fr","@gmail.com","@yahoo.fr","@hotmail.fr","@laposte.fr"
end_date = datetime.date(year=2002, month=1, day=1)

# Liste des prenoms H pour détection de la civilité
file_path_h = "hommes.txt"

with open(file_path_h, 'r', encoding='utf-8') as file:
    list_surname_men = file.read()
    list_surname_men = unidecode.unidecode(list_surname_men.lower(), "utf-8")




# Lancer une fenêtre google chrome et définir le nombre de votes à envoyer
#------------------------------------------------------------------------------------------------------
page = webdriver.Chrome(ChromeDriverManager().install())
qty_votes = 10

# Voter pour Marie
#------------------------------------------------------------------------------------------------------
for i in range(0,qty_votes): 
    page.get('https://groupemarieclaire.qualifioapp.com/20/E614CF04-A50C-4CD5-AE3C-C4C2F07D21CD/s200/v1.cfm?id=E614CF04-A50C-4CD5-AE3C-C4C2F07D21CD&pdomain=https://www.marieclaire.fr')
    participer = page.find_element(By.XPATH,'//*[@id="jouer"]')
    participer.click()

    #Choix Marie
    select_marie = page.find_element(By.XPATH,'//*[@id="answerVPic_6811060"]')
    select_marie.click()

    confirmer = page.find_element(By.XPATH,'//*[@id="suivant1"]')
    confirmer.click()


    # Remplir formulaire                                         
    #------------------------------------------------------------------------------------------------------

    # Créer un faux nom et un faux prénom (nécessaire avant la civilité)
    faux_nom_prenom = fake.name()
    print("fausse personne = ",faux_nom_prenom)
    nom = faux_nom_prenom.split()[-1]
    prenom = faux_nom_prenom.split()[0]
    
    # Renseigner une civilité cohérente dans la liste déroulante
    civ = get_civilite(prenom)
    element = page.find_element(By.ID,"choix_2048862") 
    civilite = Select(element)
    civilite.select_by_index(civ)

    # Renseigner prénom et nom dans le champ textuel
    champ_prenom = page.find_element(By.ID,"choix_2048861")
    champ_prenom.send_keys(prenom)
    champ_nom = page.find_element(By.ID,"choix_2048863")
    champ_nom.send_keys(nom)

    # Créer et renseigner fausse adresse (adresse, cp, ville)
    f = fake.address()
    address = f.split('\n')[0]
    champ_adresse = page.find_element(By.ID,"choix_2048868")
    champ_adresse.send_keys(address)

    tampon = f.split("\n")[1]
    cp = tampon.split(" ")[0]
    champ_cp = page.find_element(By.ID,"choix_2048860")
    champ_cp.send_keys(cp)

    ville = tampon.split(" ")[1]
    champ_ville = page.find_element(By.ID,"choix_2048866")
    champ_ville.send_keys(ville)

    # Créer et renseigner date de naissance aléatoire mais cohérente
    birth = fake.date_between(start_date='-50y', end_date=end_date)
    birth = birth.strftime("%d/%m/%Y")
    champ_date_naiss = page.find_element(By.ID,"2048867")
    champ_date_naiss.send_keys(birth)

    # Créer et renseigner un email suivant un pattern semi-aléatoire
    random_domain = random.choice(domain_list)

    nom_e = unidecode.unidecode(nom.lower(),"utf-8")
    prenom_e = unidecode.unidecode(prenom.lower(),"utf-8")
    email = f'{prenom_e}.{nom_e}@{random_domain}'
    champ_email = page.find_element(By.ID,"email")
    champ_email.send_keys(email)

    # Créer et renseigner un numéro mobile français
    tel = fakephone()
    champ_tel = page.find_element(By.ID,"choix_2048871")
    champ_tel.send_keys(tel)

    # Valider la saisie
    send = page.find_element(By.ID,"register")
    send.click()
```
